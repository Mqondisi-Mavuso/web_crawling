```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

#get currency 
in_currency = str(input().upper())

#get the amount in Taiwan Dollars
amount = float(input())

tables = pd.read_html('https://rate.bot.com.tw/xrt?Lang=en-US')
table = tables[0]
# Select only the first five columns of the table using the iloc function:
table = table.iloc[:, :5]
#Rename the columns of the table to more descriptive names:
table.columns = ["Currency", "Cash Buying", "Cash Selling", "Spot Buying", "Spot Selling"]
table.set_index('Currency')

# Find the row containing given currency in the "Currency" column
table_row = table[table['Currency'].str.contains(in_currency)]

# Get the "Cash Buying" value
if not table_row.empty:
    ex_rate = table_row.iloc[0]['Cash Buying']
    if ex_rate == "-":
        ex_rate = table_row.iloc[0]['Spot Buying']
    received_cash = round((amount/float(ex_rate)),3)
    print(f"{received_cash} {in_currency}")
else:
    print("Could not find that currency")
table
```
