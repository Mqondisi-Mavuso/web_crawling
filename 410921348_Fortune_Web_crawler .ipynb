```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx'
# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    soup.prettify().encode('utf-8')
    activities = []
    for i in range(10):
        activity_name = soup.find_all("span", {"id": f"BodyContent_gvActs_lblGv_act_name_{i}"})[0].get_text()
        num_hours = soup.find_all("span", {"id": f"BodyContent_gvActs_lblGv_xsl_check_{i}"})[0].get_text().split(" ")[-1]
        if num_hours.isdigit() is False:
            num_hours = 0
        else:
            num_hours = num_hours            
        num_reg = soup.find_all("span", {"id": f"BodyContent_gvActs_lblGv_reg_num_{i}"})[0].get_text().split(" ")[0]
        reg_date = soup.find_all("span", {"id": f"BodyContent_gvActs_lblGv_reg_dt_{i}"})[0].get_text()
        act_time = soup.find_all("span", {"id": f"BodyContent_gvActs_lblGv_act_dt_{i}"})[0].get_text()
    
        activities.append({
          'Name of Activity': activity_name,
          'No. of Hours Certified': num_hours,
          'No. of people Who registered': num_reg,
          'Data and Time of Registration': reg_date,
          'Data and time of Activity': act_time
        })
else: 
    print("Could not access url")
        
df = pd.DataFrame(activities)
display(df)
```
